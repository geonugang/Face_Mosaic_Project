```python
# 이미지 중복 제거 

import os

def remove_duplicate_labels(label_file):
    with open(label_file, 'r') as file:
        lines = file.readlines()
    
    # 각 레이블을 set에 추가하여 중복을 제거
    unique_lines = list(set(lines))
    
    if len(lines) != len(unique_lines):
        print(f"Removed duplicates in {label_file}")
    
    with open(label_file, 'w') as file:
        file.writelines(unique_lines)

def check_and_fix_labels(label_dir):
    for root, _, files in os.walk(label_dir):
        for file in files:
            if file.endswith('.txt'):
                label_file = os.path.join(root, file)
                remove_duplicate_labels(label_file)

# Train과 Val 레이블 디렉토리를 지정합니다.
train_label_dir = '/Users/gw/Desktop/python/face_mosaic_project/dataset/labels/train'
val_label_dir = '/Users/gw/Desktop/python/face_mosaic_project/dataset/labels/val'

# 중복된 레이블을 제거합니다.
check_and_fix_labels(train_label_dir)
check_and_fix_labels(val_label_dir)
```

```python
# 라벨 중복 제거

import os

def remove_duplicate_labels(label_file):
    with open(label_file, 'r') as file:
        lines = file.readlines()
    
    unique_lines = list(set(lines))
    
    if len(lines) != len(unique_lines):
        print(f"Removed duplicates in {label_file}")
    
    with open(label_file, 'w') as file:
        file.writelines(unique_lines)

def check_and_fix_labels(label_dir):
    for root, _, files in os.walk(label_dir):
        for file in files:
            if file.endswith('.txt'):
                remove_duplicate_labels(os.path.join(root, file))

label_dir = '/Users/gw/Desktop/python/face_mosaic_project/dataset/labels/train'
check_and_fix_labels(label_dir)

label_dir = '/Users/gw/Desktop/python/face_mosaic_project/dataset/labels/val'
check_and_fix_labels(label_dir)
```

```python
# 비정상적인 좌표값 수정

def normalize_coordinates(label_file):
    with open(label_file, 'r') as file:
        lines = file.readlines()
    
    valid_lines = []
    for line in lines:
        parts = line.strip().split()
        class_id = parts[0]
        coords = list(map(float, parts[1:]))
        
        if all(0 <= coord <= 1 for coord in coords):
            valid_lines.append(line)
        else:
            print(f"Invalid coordinates found in {label_file}: {line.strip()}")
    
    with open(label_file, 'w') as file:
        file.writelines(valid_lines)

def check_and_normalize_labels(label_dir):
    for root, _, files in os.walk(label_dir):
        for file in files:
            if file.endswith('.txt'):
                normalize_coordinates(os.path.join(root, file))

label_dir = '/Users/gw/Desktop/python/face_mosaic_project/dataset/labels/train'
check_and_normalize_labels(label_dir)

label_dir = '/Users/gw/Desktop/python/face_mosaic_project/dataset/labels/val'
check_and_normalize_labels(label_dir)
```
